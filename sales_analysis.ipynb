# Sales Store Data Analysis - Beginner Project

In this project, we'll analyze sales data to find:
- Total sales by month
- Best-selling products
- Sales trends over time
- Which regions perform best

## Step 1: Import Libraries

These libraries help us work with data and create visualizations

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# This makes plots look prettier
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

## Step 2: Load the Data

We'll use a sample sales dataset. You can download it or create one.

```python
# Load data from CSV file
# Make sure 'sales_data.csv' is in the same folder as this notebook
df = pd.read_csv('sales_data.csv')

# Display first 5 rows to see what the data looks like
print("First 5 rows of our data:")
print(df.head())

print("\nData shape (rows, columns):")
print(df.shape)

print("\nColumn names:")
print(df.columns.tolist())

print("\nData types:")
print(df.dtypes)
```

## Step 3: Explore the Data (Data Understanding)

Get basic statistics about the data

```python
# Get summary statistics
print("Summary Statistics:")
print(df.describe())

# Check for missing values
print("\nMissing values:")
print(df.isnull().sum())
```

## Step 4: Clean the Data (Data Cleaning)

Handle any missing or incorrect data

```python
# Remove any rows with missing values
df = df.dropna()

# Remove duplicate rows if any
df = df.drop_duplicates()

print(f"Data cleaned! Shape is now: {df.shape}")
```

## Step 5: Insight 1 - Total Sales by Product

Which products generate the most revenue?

```python
# Calculate total sales by product
sales_by_product = df.groupby('Product')['Sales'].sum().sort_values(ascending=False)

print("\nTotal Sales by Product:")
print(sales_by_product)

# Visualize with a bar chart
plt.figure(figsize=(10, 6))
sales_by_product.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Total Sales by Product', fontsize=16, fontweight='bold')
plt.xlabel('Product', fontsize=12)
plt.ylabel('Sales ($)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Step 6: Insight 2 - Sales by Region

Which region has the highest sales?

```python
# Calculate total sales by region
sales_by_region = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

print("\nTotal Sales by Region:")
print(sales_by_region)

# Visualize with a pie chart
plt.figure(figsize=(8, 8))
sales_by_region.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Sales Distribution by Region', fontsize=16, fontweight='bold')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

## Step 7: Insight 3 - Sales Over Time

Are sales increasing or decreasing over time?

```python
# Convert Date column to datetime (if not already)
df['Date'] = pd.to_datetime(df['Date'])

# Extract month and year
df['YearMonth'] = df['Date'].dt.to_period('M')

# Calculate monthly sales
monthly_sales = df.groupby('YearMonth')['Sales'].sum()

print("\nMonthly Sales:")
print(monthly_sales)

# Visualize with a line chart
plt.figure(figsize=(12, 6))
monthly_sales.plot(kind='line', marker='o', color='green', linewidth=2, markersize=8)
plt.title('Sales Trend Over Time', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Sales ($)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Step 8: Insight 4 - Average Order Value

What's the typical sale amount?

```python
# Calculate average sale
avg_sale = df['Sales'].mean()
median_sale = df['Sales'].median()
min_sale = df['Sales'].min()
max_sale = df['Sales'].max()

print(f"\nAverage Sale Value: ${avg_sale:.2f}")
print(f"Median Sale Value: ${median_sale:.2f}")
print(f"Minimum Sale: ${min_sale:.2f}")
print(f"Maximum Sale: ${max_sale:.2f}")

# Visualize with a histogram
plt.figure(figsize=(10, 6))
plt.hist(df['Sales'], bins=30, color='coral', edgecolor='black', alpha=0.7)
plt.axvline(avg_sale, color='red', linestyle='--', linewidth=2, label=f'Average: ${avg_sale:.2f}')
plt.axvline(median_sale, color='blue', linestyle='--', linewidth=2, label=f'Median: ${median_sale:.2f}')
plt.title('Distribution of Sale Values', fontsize=16, fontweight='bold')
plt.xlabel('Sale Amount ($)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend()
plt.tight_layout()
plt.show()
```

## Step 9: Summary & Key Findings

Let's create a summary of our analysis

```python
print("="*60)
print("SALES DATA ANALYSIS - KEY FINDINGS")
print("="*60)

print(f"\n📊 Total Records: {len(df)}")
print(f"💰 Total Sales: ${df['Sales'].sum():,.2f}")
print(f"📈 Average Sale: ${df['Sales'].mean():,.2f}")

print(f"\n🏆 Best Selling Product: {sales_by_product.index[0]} (${sales_by_product.iloc[0]:,.2f})")
print(f"🌍 Top Region: {sales_by_region.index[0]} (${sales_by_region.iloc[0]:,.2f})")

print(f"\n📅 Data Period: {df['Date'].min().date()} to {df['Date'].max().date()}")

print("\n✅ Analysis Complete!")
print("="*60)
```
